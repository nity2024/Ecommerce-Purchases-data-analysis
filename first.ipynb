# Top 5 Most Popular Email Provider (e.g. gmail.com, yahoo.com,etc...)

```python
list1=[]
for email in data['Email']:
    list1.append(email.split('@')[1])
```

```python
data['temp']=list1
```

```python
data.head(1)
```

```python
data['temp'].value_counts().head()
```

```python
data['Email'].apply(lambda x:x.split('@')[1]).value_counts().head()
```

```python
import matplotlib.pyplot as plt

# Extract top 5 email providers
top_providers = data['Email'].apply(lambda x: x.split('@')[1]).value_counts().head()

# Plot
plt.figure(figsize=(8, 5))
top_providers.plot(kind='bar', color='skyblue')
plt.title('Top 5 Email Providers')
plt.xlabel('Email Provider')
plt.ylabel('Number of Users')
plt.grid(True, axis='y')
plt.show()
```
